# 2. Data Cleaning

## 2.a Manipulasi kolom

```python
# Ubah nama kolom
df.columns = df.columns.str.strip().str.lower().str.replace(' ','_')
df.columns
```

```python
df.head()
```

```python
# hapus kolom yang tidak di perlukan
df.drop(columns=['month_number','month_name','year'], inplace=True)
```

```python
# marubah nama kolom 'date'  jadi tanggal
df.rename(columns={'date':'tanggal'},inplace=True)
```

## 2.b Hapus data inkonsisten dan merubah tipe data


```python
# data yang belum diketahuai minus atau bukan
data_minus = df.loc[df['profit'].str.contains('\(.*\)', regex=True)]
data_minus.tail(10)
```

- hapus tanda yang tidak di butuhkan 

```python
# hapus tanda yang tidak dibutuhkan
df[df.columns[4:12]] = df[df.columns[4:12]].replace(to_replace=r'[ ()$,]',value='',regex=True)

# hapus tanda - jadi 0
df["discounts"] = df["discounts"].str.replace(pat='-', repl='0', regex=True)
df['profit'] = df['profit'].str.replace(pat='-', repl='0',regex=True)
```

```python
df.head()
```

- konversi tipe data

```python
# konversi data yang bersifat currency jadi float
df[df.columns[5:12]] = df[df.columns[5:12]].astype(float)

```

```python
# konversi tanggal
df['tanggal'] = pd.to_datetime(df['tanggal'], format='%d/%m/%Y')
```

# 3. Data Normalization

## 3.a Lakukan normalisasi data yang sudah diolah sebelumnya sampai dengan 3NF

```python
df.head()
```

- Normalisasi 1NF

```python
normal_1 = df.value_counts().to_frame()
normal_1
```

```python
normal_1[normal_1.columns].nunique()
```

```python
# hapus kolom attribut turunan
df.drop(['gross_sales','discounts','sales','cogs'],axis=1, inplace=True)
```

```python
df.head()
```

- Normalisasi 2NF

```python
# check produt berdasarkan segment country, discount_band, manufakturing_price,dan date
normal_2 = df.groupby(['segment','country','discount_band','manufacturing_price','product','tanggal','units_sold'])['profit'].count().to_frame()
normal_2
```

```python
normal_2[normal_2.columns].nunique()
```

### Buat table segment

```python
# segment
segID = [f'S{11+i}' for i in range(len(df[df.columns[0]].drop_duplicates()))] # seg_id
segmentTable= pd.DataFrame(data = {
    'seg_id':segID,
    'segment':(df[df.columns[0]].drop_duplicates()).copy()
    }
)
segmentTable
```

### Buat table country

```python
# Country
c_id = [f'C{101+i}' for i in range(len(df[df.columns[1]].drop_duplicates()))] # c_id
countryTable = pd.DataFrame(data = {
    'country_id':c_id,
    'country':(df[df.columns[1]].drop_duplicates()).copy()
    }
)
countryTable
```

### Buat table discount_band

```python
# discount_band
disb_id = [f'D{41+i} 'for i in range(len(df[df.columns[3]].drop_duplicates()))] # disb_id
discount_bandTable = pd.DataFrame(data = {
    'disb_id':disb_id,
    'discount_band':(df[df.columns[3]].drop_duplicates()).copy()
    }
)
discount_bandTable
```

### Buat table Product

```python
# product dan manufacturing price
pro_id = [f'P{11+i}' for i in range(len(df[['product', 'manufacturing_price']].drop_duplicates()))] # pro_id
productTable = pd.DataFrame(data = {
    'pro_id':pro_id,
    'product':(df['product'].drop_duplicates()).copy(),
    'manufacturing_price':(df['manufacturing_price'].drop_duplicates()).copy()
    }
)
productTable
```

- Normalisasi 3NF

```python
# membuat table transection
table_transection = df[['sale_price','units_sold','profit','tanggal']]
# buat index referensi
series = [i for i in range(len(table_transection))] 
# tambah id transection
table_transection.insert(loc=0, column='trans_id', value=series)


# tabel bisa di pakai jika ada order
table_transection
```

# Buat main table 

```python
# buat mainTable
mainTable = df[df.columns[0:4]]

# index masing - masing table
segID=[]
c_id=[]
disb_id=[]
pro_id=[]

# masukan index id ke dalam list dengan cara mengecek setiap value
for i in range(len(mainTable)):
    for j in range(len(segmentTable)):
        if mainTable.iloc[i,0] == segmentTable.iloc[j,1]:
            segID.append(segmentTable.iloc[j,0])
            break
    for j in range(len(countryTable)):
        if mainTable.iloc[i,1] == countryTable.iloc[j,1]:
            c_id.append(countryTable.iloc[j,0])
            break

    for j in range(len(productTable)):
        if mainTable.iloc[i,2] == productTable.iloc[j,1]:  
            pro_id.append(productTable.iloc[j,0])        
            break

    for j in range(len(discount_bandTable)):
        if mainTable.iloc[i,3] == discount_bandTable.iloc[j,1]:  
            disb_id.append(discount_bandTable.iloc[j,0])          
            break

# Mengganti nama kolom menjadi 'trans_id'
mainTable = mainTable.copy()

# Menyisipkan kolom-kolom baru
mainTable.insert(0, column='seg_id', value=segID)
mainTable.insert(1, column='country_id', value=c_id)
mainTable.insert(2, column='pro_id', value=pro_id)
mainTable.insert(3, column='disb_id', value=disb_id) 

mainTable
```

```python
# test tabel relasi
test = pd.concat([mainTable,table_transection], axis=1)
test
```

```python
table_main = test.copy()
# karena sudah ada inisialisasi jadi drop
table_main.drop(['segment','country','product','discount_band'],axis=1, inplace=True)
```
